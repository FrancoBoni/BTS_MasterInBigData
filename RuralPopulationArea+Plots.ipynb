```python
import pandas as pd
from statistics import mode
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("rural_population_data.csv")
df.head()
```

```python
plt.figure(figsize=[10,12])
sns.barplot('Continent','LandArea', data=df)
plt.show()
```

#### Findings: Asia has a lot of rural area (at least twice as anyone else) whearas Europe and Oceania have extremely much less

```python
plt.figure(figsize=[20,15])
sns.swarmplot(x="LandArea", y="ElevationZone", data=df)
plt.show()
```

#### Findings: The only way to have large land areas is when elevation is less than 20 or Non-Contageous

```python
sns.factorplot("Continent", "Population1990", "UrbanRuralDesignation", size=4, aspect=2, data=df, kind="bar", palette="muted", legend=True)
plt.show()
```

#### Findings: The Population in 1990 was distributed towards Urban in Europe, America and South America whereas iRural n Asia and Africa.

```python
sns.factorplot("Continent", "Population2010", "UrbanRuralDesignation", size=4, aspect=2, data=df, kind="bar", palette="muted", legend=True)
plt.show()
```

#### Findings: The population in 2010 has decreased in Africa and Oceania but in Asia it has grown. In every continent, distribution seems to be cnosistent
