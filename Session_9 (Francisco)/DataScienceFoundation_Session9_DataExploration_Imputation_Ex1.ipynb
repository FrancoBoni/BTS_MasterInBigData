## 2. Are there NA Values, Outliers or Other Strange Values? Explore the variable "temp".

```python
df.isnull().sum()
```

```python
df["temp"].describe()
```

```python
df.hist(column='temp',    
                   figsize=(9,6),   
                   bins=20)         
```

```python
df["temp"].plot(kind="box",
                           figsize=(9,9))
```

```python
new_age_var = np.where(df["temp"].isnull(),    # Logical check
                       random.randint(14,26),  # If True, a random number between the 1Q & 3Q will be choosen
                       df["temp"])             # Value if check is false

df["New Temperature"] = new_age_var 

df["New Temperature"].describe()
```

```python
df.hist(column='New Temperature',    
                   figsize=(9,6),   
                   bins=20)     
```

```python
df["New Temperature"].plot(kind="box",
                           figsize=(9,9))
```

```python
df.corr()
```

En este cuadro tenemos que considerar que hay una correlacion cuando un valor supera 0,65.

```python
plt.figure(figsize=[14,8])
sns.heatmap(df.corr())
```

Let's see the values of the correlatons from the heatmap:

```python
df.drop(['temp'], axis=1).corr(method='spearman').style.format("{:.2}").background_gradient(cmap=plt.get_cmap('bwr'), axis=1)
```

This heatmap show a high correlation between the data. The variable temp has a good (which in this case is New Temp as we previously stated) has a very tight correlation with atemp (that is the thermal sensation). This is logic as atemp is mainly based on the temperature plus some other factors such as wind, sunlight, humidity, etc.
